```r
# google file ID
data<-read.csv("https://raw.githubusercontent.com/rushidarge/Rlab_College/main/data/home.csv")
data
```

```r
# a. print the 5 point statistical summary for any 2 numeric attributes from your data set
fivenum(data$MSSubClass) 
fivenum(data$LotFrontage)
```

```r
# b. select 2 atributes have closest mean value and draw a scatter plot with proper x and labelsloi
# ntt and title (for plotting you must use R library functions or gnuplot and not matplotlib )
means<- colMeans(data[sapply(data, is.numeric)] )
means

library(ggplot2)

MoSold<-data$MoSold
LotFrontage<-data$LotFrontage
plot(data$MoSold , data$LotFrontage , main = "Scatterplot" , xlab = "MSSubClass" , ylab = "LotFrontage" , las = 1 , xlim = c(0,2) , col = 2 )
```

```r
# e. Write a R function to plot histogram of some specific input attribute for any data set provided
new.function <- function(LotFrontage) {
   hist(LotFrontage)
}

new.function(LotFrontage)
```

```r
# c. Find coefficient of correlation using some R library function between above 2 attributes
cor(LotFrontage,MoSold, method = c("pearson", "kendall", "spearman"))
```

```r
boxplot(LotFrontage,MoSold)
title("LotFrontage vs. MoSold")
```
